```python
import random
import time

def insert_sort(lst):
    if len(lst) < 2: return lst
    
    for i in range(1, len(lst)):
        j = i - 1
        key = lst[i]
        while j >= 0:
            if lst[j] > key:
                lst[j+1] = lst[j]
                j -= 1
            else:
                break
        lst[j+1] = key
    return lst

def selection_sort(lst):
    for i in range(len(lst)):
        min_idx = i
        for j in range(i + 1, len(lst)):
            if lst[j] < lst[min_idx]:
                min_idx = j
        lst[i], lst[min_idx] = lst[min_idx], lst[i]
    return lst

def partition(arr, p, r):
    x = arr[r]
    i = p - 1
    for j in range(p, r):
        if arr[j] <= x:
            i = i + 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i+1], arr[r] = arr[r], arr[i+1]
    return i+1
       
def quicksort(arr, p, r):
    stack = [(p, r)]
    while stack:
        p, r = stack.pop()
        if p < r:
            q = partition(arr, p, r)
            stack.append((p, q-1))
            stack.append((q+1, r))
    
# 生成随机数列
def generate_list(n):
    return [random.randint(1, 10000) for _ in range(n)]

# 生成三组随机数列
lst1 = generate_list(100)
lst2 = generate_list(1000)
lst3 = generate_list(10000)

#第一组
print("第一组：")
start_time = time.time()
insert_sort(lst1)
end_time = time.time()
print("插入排序:", end_time - start_time, "秒")
start_time = time.time()
selection_sort(lst1)
end_time = time.time()
print("选择排序:", end_time - start_time, "秒")
start_time = time.time()
quicksort(lst1, 0, len(lst1)-1)
end_time = time.time()
print("快速排序:", end_time - start_time, "秒")

#第二组
print("\n第二组：")
start_time = time.time()
insert_sort(lst2)
end_time = time.time()
print("插入排序:", end_time - start_time, "秒")
start_time = time.time()
selection_sort(lst2)
end_time = time.time()
print("选择排序:", end_time - start_time, "秒")
start_time = time.time()
quicksort(lst2, 0, len(lst2)-1)
end_time = time.time()
print("快速排序:", end_time - start_time, "秒")

#第三组
print("\n第三组：")
start_time = time.time()
insert_sort(lst3)
end_time = time.time()
print("插入排序:", end_time - start_time, "秒")
start_time = time.time()
selection_sort(lst3)
end_time = time.time()
print("选择排序:", end_time - start_time, "秒")
start_time = time.time()
quicksort(lst3, 0, len(lst3)-1)
end_time = time.time()
print("快速排序:", end_time - start_time, "秒")
```
